# Set Up API Request
Define the API endpoint and parameters required for the request.

```python
# Define the API endpoint
api_endpoint = "https://verkehr.autobahn.de/o/autobahn"

# Define the parameters for the request
params = {
    "format": "json"
}

# Make the API request
response = requests.get(api_endpoint, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
else:
    data = None
    print(f"Failed to retrieve data: {response.status_code}")
```

# Get Current Road Work Data



```python
def get_roadworks(autobahn_id):
    url = f"https://verkehr.autobahn.de/o/autobahn/{autobahn_id}/services/roadworks"
    response = requests.get(url)
    return response.json()['roadworks']


roadworks = get_roadworks("A1")
print(roadworks)
print(type(roadworks))
print(type(roadworks[0]))
print(roadworks[0].keys())	
print(roadworks[0]['coordinate'])
```

```python
import folium

# Create a map centered around a specific location (latitude, longitude)
m = folium.Map(location=[51.1657, 10.4515], zoom_start=6)  # Centered on Germany

# loop through all roadworks and add markers to the map
for roadwork in roadworks:
    lat = roadwork['coordinate']['lat']
    lon = roadwork['coordinate']['long']
    folium.Marker(
       location=[float(lat), float(lon)],
       popup=roadwork['description']
    ).add_to(m)


# Save the map to an HTML file
m.save("traffic_map.html")

```
